```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w1 = 0.8
w2 = 0.7

def ReLU(x):
    return x * (x > 0)

x = np.linspace(-10, 10, 100)
y1 = 1/(1 + np.exp(-(w1 + w2*x)))
y2 = ReLU(x)

fig, axs = plt.subplots(1, 2, figsize = (12, 5))
fig.suptitle('Activation Functions', fontsize=16)

axs[0].plot(x, y1)
axs[0].set_title('Logistic/Sigmoid')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')

axs[1].plot(x, y2)
axs[1].set_title('Rectified Linear Uinit (ReLU)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')
```
